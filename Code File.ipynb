# 1. Comparing the Win/Loss Record of Each Team in IPL Seasons
The visualization compares the win/loss record of each team in IPL seasons, showing the number of wins per team per season in a clustered bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouping data by season and team to get the number of wins per team per season
team_wins_by_season = df.groupby(['Season', 'winner'])['winner'].count().unstack()

# Creating clustered bar chart
fig, ax = plt.subplots(figsize=(15,5))
bar_width = 0.05
num_teams = len(team_wins_by_season.columns)
x = range(len(team_wins_by_season.index))
for i, team in enumerate(team_wins_by_season.columns):
    ax.bar([j + i * bar_width for j in x],
           team_wins_by_season[team],
           width=bar_width,
           label=team)

# Customizing chart labels and title
ax.set_xlabel('Season')
ax.set_ylabel('Number of wins')
ax.set_title('Team Performance over Seasons')

# Setting legend outside the plot area
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Setting width of bars for better visibility
ax.set_xticks([j + (num_teams-1)/2 * bar_width for j in x])
ax.set_xticklabels(team_wins_by_season.index)

# Showing the chart
plt.show()
```

# 2. Pie Chart of Percentage of Wins by Teams Batting First vs Batting Second.

The pie chart shows that in the Indian Premier League, teams batting second have a higher percentage of wins than teams batting first, with Kolkata Knight Riders having the highest percentage of wins among teams batting second, and Mumbai Indians having the highest percentage of wins among teams batting first.

```python
import matplotlib.pyplot as plt


# creating a new column to identify the team that won (either team1 or team2)
df['winning_team'] = df['team1']
df.loc[df['win_by_runs'] < 0, 'winning_team'] = df['team2']

# counting the number of matches won by each team batting first
batting_first = df[df['win_by_runs'] > 0]['winning_team'].value_counts()

# counting the number of matches won by each team batting second
batting_second = df[df['win_by_wickets'] > 0]['winning_team'].value_counts()

# combining the two counts into a single dataframe
win_counts = pd.DataFrame([batting_first, batting_second], index=['batting_first', 'batting_second'])

# plotting the pie chart
win_counts.transpose().plot(kind='pie', subplots=True, figsize=(31, 31), autopct='%1.1f%%', legend=False)
plt.title('Percentage of Matches Won by Batting First vs Batting Second')
plt.ylabel('')
plt.show()
```

# 4. Get the top player of match winner in each Venue?

The scatter plot reveals that Chris Gayle has won the player of the match award at the M Chinnaswamy Stadium in Bengaluru more times than any other player. Similarly, AB de Villiers has won the award at the same stadium on the second most occasions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the IPL matches dataset
matches = pd.read_csv('C:/Users/divya/OneDrive/Desktop/Montclair/Project/EDA/IPL.CSV')

# Grouping the dataset by venue and player_of_match, and count the number of occurrences
top_players = matches.groupby(['venue', 'player_of_match']).size().reset_index(name='counts')

# Sorting the dataset by venue and counts in descending order
top_players = top_players.sort_values(['venue', 'counts'], ascending=[True, False])

# Getting the top player of match for each venue
top_players = top_players.groupby('venue').head(1)

# Setting the size of the plot
plt.figure(figsize=(15, 8))

# Creating a scatter plot
sns.scatterplot(data=top_players, x='venue', y='player_of_match', size='counts', sizes=(100, 1000), alpha=0.7)

# Setting the title and axis labels
plt.title('Top Player of Match Winners in Each Venue')
plt.xlabel('Venue')
plt.ylabel('Player of Match')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=90)

# Showing the plot
plt.show()
```

# 2. Percentage of Matches won by Teams who won and Lost the Toss

```python
# creating a new column 'won_toss_and_won_match' to indicate if the team that won the toss also won the match
df['won_toss_and_won_match'] = (df['toss_winner'] == df['winner'])

# calculating the percentage of matches won by teams who won the toss
toss_win_match_win_pct = df[df['toss_winner']==df['winner']]['winner'].count() / df['winner'].count()

# calculating the percentage of matches won by teams who lost the toss
toss_loss_match_win_pct = df[df['toss_winner']!=df['winner']]['winner'].count() / df['winner'].count()
print("Percentage of Matches Won by Teams Who Won the Toss: ", toss_win_match_win_pct)
print("Percentage of Matches Won by Teams Who Lost the Toss: ", toss_loss_match_win_pct)

labels = ['Won Toss and Won Match', 'Lost Toss and Won Match']
sizes = [toss_win_match_win_pct, toss_loss_match_win_pct]
colors = ['pink', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Percentage of Matches Won by Teams Who Won and Lost the Toss')
plt.show()
```
